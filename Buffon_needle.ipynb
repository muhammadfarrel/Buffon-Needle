```python
import numpy as np
import matplotlib.pyplot as plt

ntot = int(input("n total ")) #total needle
#d > l
d = float(input("d ")) #jarak antara garis
l = float(input("l ")) #panjang needle
```

### Model

```python
def needle(ntot, d, l):
    hit = 0
    i = 0
    while i <= ntot:

        #random theta
        theta = np.random.uniform(0, np.pi/2)

        #random x
        x = np.random.uniform(0,d)

        #cek apakah needle berselisih dengan garis x random
        if (x <= l * np.cos(theta)):
            hit = hit + 1

        i = i + 1
        
    phi = 2*l*ntot/(d*hit)
    return phi
```

### Let's Try

```python
exp = 1000
pi_array = []
pi_average = []
for i in range(0,exp):
    pi = needle(ntot,d,l)
    pi_array.append(pi)
    if(i%10 == 0 and i != 0):
        pi_avgg = sum(pi_array[-10:]) / 10
        pi_average.append(pi_avgg)
pi_avgg = sum(pi_array[-10:]) / 10
pi_average.append(pi_avgg)
```

```python
plt.scatter(range(0,exp), pi_array, c='blue', label = 'pi percobaan', s = 10)

plt.plot(np.arange(10,exp+10,10),pi_average, c = 'red', label = 'pi avg')

plt.axhline(y = 3.14, c = 'black', label = 'pi exact')

plt.legend(loc = 'upper right')
plt.show()
```
